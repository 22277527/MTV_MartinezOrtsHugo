# 1. (2 points) We have performed a multiple linear regression to determine per capita wager in the Royal Lottery of 1830’s France using a few characteristics:


### a) **Interpretation of Results (1.5 points)**

A multiple linear regression was performed to model per capita spending on the Royal Lottery in 1830s France as a function of:

- **Literacy**
- **Wealth**
- **Region** (a categorical variable with multiple levels)

#### Key Findings:

1. **Model Fit:**
   - **R² = 0.338**: The model explains 33.8% of the variance in lottery spending.
   - **Adjusted R² = 0.287**: Slightly lower, accounting for the number of predictors.
   - **F-statistic = 6.636, p-value = 1.07e-05**: The model is statistically significant overall.

2. **Predictor Coefficients:**
   - **Intercept = 38.65**: Baseline lottery spending for the reference region (likely the omitted category), with 0 wealth and literacy (not directly interpretable).
   - **Wealth = 0.4515, p < 0.001**: Statistically significant. Each unit increase in wealth increases lottery spending by ~0.45 units.
   - **Literacy = -0.1858, p = 0.63**: Not statistically significant. Literacy does not show a clear linear effect.
   - **Region (T.E, T.N, T.S, T.W)**: None of the regional dummy variables are statistically significant (all p > 0.05), suggesting no strong regional differences compared to the reference region.

3. **Model Diagnostics:**
   - **Skew = -0.340**, **Kurtosis = 2.454**: Residuals are fairly close to normal distribution.
   - **Durbin-Watson = 1.785**: No strong evidence of autocorrelation.
   - **Omnibus p-value = 0.218**: Residuals do not significantly deviate from normality.

---

### b) **Prediction Model Justification (0.5 points)**

This model **could be used for prediction**, but it is **not ideal in its current state**:

- The **Adjusted R² is low (0.287)**, meaning much of the variance is unexplained.
- **Only Wealth is a statistically significant predictor**.
- **No assessment of multicollinearity or interactions** has been performed.
