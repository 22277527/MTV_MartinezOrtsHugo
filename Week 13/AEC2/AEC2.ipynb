# 1. (2 points) It has been proven that the waiting time (in minutes) until treatment in a particular emergency department follows a normal distribution. An average waiting time of 14.25 minutes and a standard deviation of 2.5 minutes were calculated from a sample of 100 people who attended this service.
a) Could we say, with a significance level of 5 %, that the average waiting time in this emergency department
is not 15 minutes? (zα/2 = 1.96)

b) What could we conclude if the significance level had been 0.1%? (zα/2 = 3.27)

c) Is there a contradiction in both situations?


## Analysis of Waiting Time in an Emergency Department

We aim to determine whether the average waiting time in an emergency department differs from 15 minutes, based on a sample of 100 individuals. The sample yielded a mean waiting time of 14.25 minutes and a standard deviation of 2.5 minutes. A two-tailed hypothesis test will be performed using the Z-test, since the distribution is normal and the population standard deviation is known.

### Hypotheses

- **Null hypothesis**: \( H_0: \mu = 15 \)  
- **Alternative hypothesis**: \( H_1: \mu \ne 15 \)

### Test Statistic Calculation

\[
Z = \frac{\bar{x} - \mu_0}{\sigma / \sqrt{n}} = \frac{14.25 - 15}{2.5 / \sqrt{100}} = \frac{-0.75}{0.25} = -3
\]

---

### a) Significance level of 5% ( \( z_{\alpha/2} = 1.96 \) )

The absolute value of the test statistic is \( |Z| = 3 \), which is **greater than 1.96**.  
Therefore, **we reject the null hypothesis**.

**Conclusion**: At the 5% significance level, we conclude that **the average waiting time is significantly different from 15 minutes**.

---

### b) Significance level of 0.1% ( \( z_{\alpha/2} = 3.27 \) )

The absolute value of the test statistic is \( |Z| = 3 \), which is **less than 3.27**.  
In this case, **we fail to reject the null hypothesis**.

**Conclusion**: At the 0.1% significance level, **we cannot conclude** that the average waiting time differs significantly from 15 minutes.

---

### c) Is there a contradiction between the results?

**There is no contradiction.**  
A lower significance level requires stronger evidence to reject the null hypothesis.  
A result can be statistically significant at the 5% level and not significant at the 0.1% level, which is entirely consistent within inferential statistics.

# 2 points) A small reprographic company, which has a single printer, wants to study the relationship between the printing time of a job (Y) and the number of pages of the job to print (X) to manage orders and the assignment of tasks better. For this purpose, a sample of works has been collected over a week, and for each of them, the number of pages of the work and the printing time of the same has been noted. Following some statistical analyses, the following results have been obtained





## a) Estimate the parameters of the linear regression model

The linear regression model is:

\[
Y = \beta_0 + \beta_1 X
\]

From the regression output:

- Intercept (β₀): **2.6272**
- Slope (β₁): **0.4286**

**Estimated regression equation:**

\[
\hat{Y} = 2.6272 + 0.4286 X
\]

This model estimates the printing time based on the number of pages to print.

---

## b) Goodness of fit – R² calculation and interpretation

The coefficient of determination \( R^2 \) is calculated as:

\[
R^2 = \frac{SS_{\text{Model}}}{SS_{\text{Total}}} = \frac{39343.3}{47368.9} \approx 0.8304
\]

**Interpretation:**

About **83.04%** of the variation in printing time is explained by the number of pages. This indicates a **very strong fit** of the model.

---

## c) Correlation coefficient between X and Y

\[
r = \sqrt{R^2} = \sqrt{0.8304} \approx 0.9112
\]

**Interpretation:**

There is a **strong positive linear correlation** between the number of pages and the printing time.

---

## d) Statistical significance of parameters and regression equation

From the regression table:

- Intercept p-value = 0.0000
- Num Pages slope p-value = 0.0000

At \( \alpha = 0.05 \), both parameters are **statistically significant**.

**Final regression equation:**

\[
\hat{Y} = 2.6272 + 0.4286 X
\]

---

## e) Expected printing time for a 6-page job

Using the model:

\[
\hat{Y} = 2.6272 + 0.4286 \cdot 6 = 2.6272 + 2.5716 = \boxed{5.1988 \text{ minutes}}
\]

---

## f) Residual variance and interpretation

\[
\text{Residual Variance} = \frac{SS_{\text{Residuals}}}{df_{\text{Residuals}}} = \frac{8025.61}{73} \approx \boxed{109.94}
\]

**Interpretation:**

This value represents the average unexplained variability in the printing time. Although the model explains most of the variation, a residual variance of 109.94 shows that some variation remains due to other factors or randomness.
