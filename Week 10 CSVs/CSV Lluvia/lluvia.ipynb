```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Default Dataset.csv', sep= ";")
display(df.head(10))
```

```python
# Suponiendo que el CSV tiene columnas 'Fecha' y 'Valor'
fig = go.Figure()

# Agregar la serie de datos
fig.add_trace(go.Scatter(
    x=df['Fecha'], 
    y=df['Valor'], 
    mode='lines+markers', 
    marker=dict(color='blue', size=4),
    line=dict(color='blue', width=1),
    name="Datos"
))

# Agregar una línea horizontal en y=10 (ajusta si es necesario)
fig.add_trace(go.Scatter(
    x=df['Fecha'], 
    y=[10]*len(df),  
    mode='lines',
    line=dict(color='red', width=2),
    name="Referencia"
))

# Ajustes de diseño
fig.update_layout(
    title="Gráfica de Datos",
    xaxis_title="Fecha",
    yaxis_title="Valor",
    xaxis=dict(showgrid=True),
    yaxis=dict(showgrid=True),
    template="plotly_white"
)

# Mostrar la gráfica
fig.show()
```
